```python
def graph_v1(dataframe, title):
    """
    Draw a graph, figure data is appended into a list
    """
    INCREASING_COLOR = '#17BECF'
    DECREASING_COLOR = '#7F7F7F'
    
    colors = []
    for i in range(len(dataframe.Close)):
        if i != 0:
            if dataframe.Close[i] > dataframe.Close[i-1]:
                colors.append(INCREASING_COLOR)
            else:
                colors.append(DECREASING_COLOR)
        else:
            colors.append(DECREASING_COLOR)
            
    data = []
        
    data.append(go.Candlestick(x = dataframe.index,
                           open = dataframe.Open,
                           high = dataframe.High,
                           low = dataframe.Low,
                           close = dataframe.Close,
                           increasing = dict( line = dict( color = INCREASING_COLOR ) ),
                           decreasing = dict( line = dict( color = DECREASING_COLOR ) ),
                           yaxis = 'y2', name = 'Candlestick'))

    data.append(go.Bar(x = dataframe.index, y = dataframe.Volume,
                       marker = dict( color = colors ),
                       yaxis = 'y', name = 'Volume'))

    layout = go.Layout(yaxis = dict(domain = [0, 0.3], showticklabels = False),
                       yaxis2 = dict(domain = [0.7, 1.0]), title = title)
    
    
    fig = go.Figure(data = data, layout = layout)
    fig.update_layout(
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                         label="1m",
                         step="month",
                         stepmode="backward"),
                    dict(count=6,
                         label="6m",
                         step="month",
                         stepmode="backward"),
                    dict(count=1,
                         label="YTD",
                         step="year",
                         stepmode="todate"),
                    dict(count=1,
                         label="1y",
                         step="year",
                         stepmode="backward"),
                    dict(step="all")
                ])
            ),
            rangeslider=dict(
                visible=True
            ),
            type="date"
        )
    )
    fig.show()
```

```python
def graph_v2(dataframe, title):
    """
    Draw a graph, figure data is added with traces
    """
    INCREASING_COLOR = '#17BECF'
    DECREASING_COLOR = '#7F7F7F'
    
    colors = []
    for i in range(len(dataframe.Close)):
        if i != 0:
            if dataframe.Close[i] > dataframe.Close[i-1]:
                colors.append(INCREASING_COLOR)
            else:
                colors.append(DECREASING_COLOR)
        else:
            colors.append(DECREASING_COLOR)
        
    layout = go.Layout(
        yaxis1 = dict(
            domain = [0, 0.4], 
            title="Volume", 
            side="right",
            tickmode = "array",
            tickvals = [30,50,70,90],
            ticktext = [30,50,70,90]
        ),
        yaxis2 = dict(domain = [0.4, 1.0], title="Value"), 
        title = title)
    fig = go.Figure(layout = layout)
    
    fig.add_trace(
        go.Candlestick(
            x = dataframe.index,
            open = dataframe.Open,
            high = dataframe.High,
            low = dataframe.Low,
            close = dataframe.Close,
            increasing = dict( line = dict( color = INCREASING_COLOR ) ),
            decreasing = dict( line = dict( color = DECREASING_COLOR ) ),
            yaxis = 'y2', name = 'Candlestick'
        )
    )
    
    fig.add_trace(
        go.Bar(
            x = dataframe.index, 
            y = dataframe.Volume,
            marker = dict( color = colors ),
            yaxis = 'y1', name = 'Volume'
        )
    )
    
    fig.update_layout(
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=7,
                        label="7d",
                        step="day",
                        stepmode="backward"),
                    dict(count=1,
                         label="1m",
                         step="month",
                         stepmode="backward"),
                    dict(count=6,
                         label="6m",
                         step="month",
                         stepmode="backward"),
                    dict(count=1,
                         label="YTD",
                         step="year",
                         stepmode="todate"),
                    dict(count=1,
                         label="1y",
                         step="year",
                         stepmode="backward"),
                    dict(step="all")
                ])
            ),
            rangeslider=dict(
                visible=True
            ),
            type="date"
        )
    )
    fig.show()
```

```python
hourly = hourly[:4000]
graph_v2(hourly, "Hourly volumes and values")
```

```python
graph_v2(monthly, "Monthly volumes and values")
```

```python
graph_v2(weekly, "Weekly volumes and values")
```
